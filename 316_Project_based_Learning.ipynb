```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

# 1.Density Plot

1. Create a density plot for at least one numerical variable

```python
# Density plot for 'sepal width (cm)'
df['sepal width (cm)'].plot.kde()
plt.title('Density Plot of Sepal Width (cm)')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

2.Create multiple density plots (overlay) for comparison (e.g., by category)\
3. Label axes, title, and legend clearly

```python
#Density Plot
ax = df.plot.kde(bw_method='scott', linestyle='-', linewidth=2, figsize=(10, 6))
plt.title('Density Plots of Iris Features')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(title='Features')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Plot density for all columns
df.plot.density(bw_method='silverman', figsize=(10, 6))
```

using Seaborn and matplotlib

```python
import seaborn as sns

df['species'] = data.target_names[data.target]
sns.kdeplot(df, x='petal length (cm)', hue='species', fill=True)
plt.title('Density Plot of Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.show()
```

**1.What the density plot shows**\
->Imagine you have a bunch of measurements, like the petal lengths of different flowers. A density plot, also called a Kernel Density Estimate (KDE) plot, is like a super-smooth histogram that shows you where those measurements tend to pile up and how spread out they are.

Think of it as drawing a smooth curve that outlines the shape of the data's distribution.

Where the curve is high (peaks): That's where most of your measurements are concentrated.
Where the curve is low (valleys): That's where you have fewer measurements.
The overall shape: Tells you if the data is symmetric, skewed (lopsided), or if it has multiple peaks (meaning there might be distinct groups within your data).
Essentially, it gives you a clear visual idea of the probability of finding a measurement at any given value. It's great for understanding the underlying distribution of a single numerical variable.

**2.Where data is concentrated**\
->When we talk about 'where data is concentrated' in a density plot, we're essentially looking for the peaks or highest points on the curve.

Imagine the density plot is a mountain range. The highest mountain peaks represent the values where the data points are most numerous or occur most frequently. If you were to randomly pick a data point, it's most likely to be found in these 'densely populated' areas.

Conversely, the valleys or flatter parts of the curve show where data is less concentrated, meaning those values appear less often in your dataset.



# 2.Contour Plot

```python
sns.kdeplot(df, x='petal length (cm)', y='petal width (cm)', fill=True)
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], alpha=0.5)
plt.title('2d Density Plot of Petal Length and Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width')
plt.grid(True,linestyle='--',alpha=0.6)
plt.show()
```

**3.Any patterns, clusters, or outliers**\

-->Analyzing the density and contour plots of the Iris dataset reveals several interesting patterns and clusters:

Single Variable Density Plots (e.g., Sepal Width): While some individual feature density plots (like 'sepal width (cm)') might show slight bimodality or skewness, hinting at different underlying groups, they don't explicitly separate into clear clusters on their own.

Density Plot of Petal Length by Species (Seaborn KDE plot with hue='species'): This plot is a strong indicator of clustering based on species:

Setosa: Shows a very distinct and separate peak at lower petal lengths, indicating a clear cluster with smaller petals.
Versicolor: Occupies a middle range of petal lengths, forming another distinct cluster.
Virginica: Exhibits a peak at higher petal lengths, forming a third cluster. There might be some overlap between Versicolor and Virginica distributions, but they are generally well-separated.
2D Density Plot of Petal Length and Petal Width (Contour Plot): This plot visually confirms the clustering observed in the species-separated density plot and reveals more about their relationships:

You can clearly see three distinct high-density regions (clusters) in the 2D space, each corresponding to an Iris species. These clusters are quite well-separated, especially the Setosa cluster from the other two.
There's a strong positive correlation between 'petal length (cm)' and 'petal width (cm)'. As petal length increases, petal width generally increases across all species.
Outliers: While the plots don't show extremely isolated points far from all data, any individual scattered points or regions with very low contour density far from the main clusters could be considered potential outliers, or less typical observations within their respective species groups. However, the data largely congregates within these three well-defined clusters.

**4.Insights from contour plot**\
The contour plot (or 2D density plot) is particularly insightful because it visually brings together two related variables – in this case, 'petal length (cm)' and 'petal width (cm)'. Think of it like a topographic map for your data!

Here's what it tells us:

Clear Groupings: You can distinctly see three 'mountains' or high-density regions. Each of these 'mountains' represents a cluster of data points, and we know from our earlier analysis that these correspond to the three different Iris species (setosa, versicolor, and virginica). The more isolated these 'mountains' are, the more distinct the groups.

Relationships Between Variables: The way the contours stretch and align shows a strong relationship. For the Iris data, you'll notice that as petal length increases, petal width generally increases too. This suggests a positive correlation between these two features.

Overlap and Separation: The plot highlights areas where species might overlap (where the contours from different 'mountains' get close) and where they are clearly separated (like the setosa cluster, which is quite isolated). This helps us understand how easy it would be to distinguish between species based on these two features alone.

Essentially, the contour plot is a fantastic way to see how two numerical variables interact and whether they form distinct groups within your dataset.
